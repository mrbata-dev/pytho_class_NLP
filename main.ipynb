```python
# 2 ways to handling outliers
# 1. Trimming outliers-remove the rows having outliers/
#2. capping outliers

#trimming


#IQR
Q1=df['Life expectancy '].quantile(0.25)
Q3=df['Life expectancy '].quantile(0.75)
IQR=Q3-Q1
print(Q1, Q3)
print(IQR)
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
df[(df['Life expectancy ']> upper_bound ) | (df['Life expectancy ']< lower_bound)]
df
```

```python
import numpy as np
df['Life expectancy ']=np.clip(df['Life expectancy '], min=lower_bound, max=upper_bound)
df
# df['Life expectancy '].max()
```

```python
df.columns

df.drop(['infant deaths', 'percentage expenditure',  ' thinness 5-9 years', 'Income composition of resources'], axis=1)
```

```python
df.dropna()
```

```python
df['Life expectancy '].mean()
```

```python
df['Life expectancy '].fillna(df['Life expectancy '].mean(), inplace=True)
df['Adult Mortality'].fillna(df['Adult Mortality'].mean(), inplace=True)
df['Alcohol'].fillna(df['Alcohol'].mean(), inplace=True)
df['Hepatitis B'].fillna(df['Hepatitis B'].mean(), inplace=True)
df[' BMI '].fillna(df[' BMI '].mean(), inplace=True)
df['Polio'].fillna(df['Polio'].mean(), inplace=True)
df['Total expenditure'].fillna(df['Total expenditure'].mean(), inplace=True)
df['Diphtheria '].fillna(df['Diphtheria '].mean(), inplace=True)
df['GDP'].fillna(df['GDP'].mean(), inplace=True)
df['Population'].fillna(df['Population'].mean(), inplace=True)
df[' thinness  1-19 years'].fillna(df[' thinness  1-19 years'].mean(), inplace=True)
df[' thinness 5-9 years'].fillna(df[' thinness 5-9 years'].mean(), inplace=True)
df['Income composition of resources'].fillna(df['Income composition of resources'].mean(), inplace=True)
df['Schooling'].fillna(df['Schooling'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
#seperate dependent and independent variables 

X=df.drop(['Life expectancy '], axis=1)
y=df['Life expectancy ']
X.shape
y.shape
```

```python
#ML
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(X, y, test_size=0.4, random_state=44)
x_train
```

```python
df.isnull().sum()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train )
```

```python
#Predict 
y_predict = model.predict(x_test)
y_predict
```

```python
result = pd.DataFrame({'actual': y_test, 'predicted': y_predict})
result
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_predict)
```
